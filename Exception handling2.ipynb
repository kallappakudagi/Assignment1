# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
#  Lookup Error acts as a base class for the exceptions that occur when a key or index used on a mapping or sequence of a list/dictionary is invalid
#  or does not exists. The two types of exceptions raised are: IndexError. KeyError.
```

```python
## index error

x = [1, 2, 3, 4]
try:
    print(x[10])
except IndexError as e:
    print(f"{e}")
```

```python
ages = {'Jim': 30, 'Pam': 28, 'Kevin': 33}
person = input('Get age for: ')

try:
    print(f'{person} is {ages[person]} years old.')
except KeyError:
    print(f"{person}'s age is unknown.")
```
